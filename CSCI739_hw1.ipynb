```python
import numpy as np
import matplotlib.pyplot as plt 


N=1000
np.random.seed(47405)
flipsequence=np.random.choice(a=[0,1],p=[.5,.5],size=N,replace=True)
res="Flip sequence="
endProportion="End Proportion="
for index in range(10):
    if flipsequence[index]==0:
        res+="T"
    else:
        res+="H"
res+="..."

r=np.cumsum(flipsequence)
prob=[]
for i in range(len(flipsequence)):
    prob.append(r[i]/(i+1))
endProportion+=str(prob[-1])


n=[i for i in range(1,N+1,1)]
line=[0.5]*len(prob)


plt.plot(n,prob,"ob-")
plt.plot(n,line,ls="--",c="r")
plt.text(10,0.9,res)
plt.text(50,0.2,endProportion)


plt.xscale('log')
plt.xlim(1,N)
plt.ylim(0.0,1.0)
plt.xlabel("Flip Number")
plt.ylabel("Proportion Heads")
plt.title("Running Proportion of Heads")
```
